```python
from langchain_core.tools import tool
```

```python
#step 1
def add(x,y):
    """ Add two numbers """
    return x+y
```

### 1st Method using @tool decorator

```python
@tool # use this decorator to create a tool
def multiply(a: int, b:int) -> int:
    """ Multiply two numbers """
    return a * b

```

```python
result = multiply.invoke({"a":3, "b":5})
print(result)   
```

```python
print(multiply.name) # gives name of tool or function
print(multiply.description) # gives description of tool which is a comment in the fucntion of tool
print(multiply.args)
```

```python
print(multiply.args_schema.model_json_schema()) 
```

### 2nd Method using StructuredTool

```python
pip install langchain.tools.StructuredTool
```

```python
from langchain_core.tools import StructuredTool
from pydantic import BaseModel, Field
```

```python
class MultiplyInput(BaseModel):
    a: int = Field(required=True, description="The first number to add")
    b: int = Field(required=True, description="The second number to add")
```

```python
def multiply_func(a: int, b: int) -> int:
    return a * b
```

```python
multiply_tool = StructuredTool.from_function(
    func=multiply_func,
    name="multiply",
    description="Multiply two numbers",
    args_schema=MultiplyInput
)
```

```python
result = multiply_tool.invoke({'a':3, 'b':3})

print(result)
print(multiply_tool.name)
print(multiply_tool.description)
print(multiply_tool.args)
```

### Method 3 - Using BaseTool Class

```python
from langchain.tools import BaseTool
from typing import Type
```

```python
# arg schema using pydantic

class MultiplyInput(BaseModel):
    a: int = Field(required=True, description="The first number to add")
    b: int = Field(required=True, description="The second number to add")
```

```python
class MultiplyTool(BaseTool):
    name: str = "multiply"
    description: str = "Multiply two numbers"

    args_schema: Type[BaseModel] = MultiplyInput

    def _run(self, a: int, b: int) -> int:
        return a * b
```

```python
multiply_tool = MultiplyTool()
```

```python
result = multiply_tool.invoke({'a':3, 'b':3})

print(result)
print(multiply_tool.name)
print(multiply_tool.description)

print(multiply_tool.args)
```

### TOOL KIT - COLLECTION OF MULTIPLE TOOLS

```python
from langchain_core.tools import tool

# Custom tools
@tool
def add(a: int, b: int) -> int:
    """Add two numbers"""
    return a + b

@tool
def multiply(a: int, b: int) -> int:
    """Multiply two numbers"""
    return a * b
```

```python
class MathToolkit:
    def get_tools(self):
        return [add, multiply]
```

```python
toolkit = MathToolkit()
tools = toolkit.get_tools()

for tool in tools:
    print(tool.name, "=>", tool.description)
```
