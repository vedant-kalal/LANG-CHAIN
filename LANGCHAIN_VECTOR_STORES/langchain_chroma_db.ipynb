```python
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_chroma import Chroma
```

```python
from typing_extensions import Doc
from langchain_core.documents import Document

doc1 = Document(
    page_content="cristiano ronaldo is best football player in the world and has highest goal in football history, his shooting speed in very high",
    metadata={"team" : "real madrid"}
)

doc2 = Document(
    page_content="lionel messi is 2nd best football player in the world and has highest assists in football history, he dribbles so smoothly",
    metadata={"team" : "barcelona"}
)

doc3 = Document(
    page_content="neymar jr is 3rd best football player in the world, he has best skills among football players,his footwork is so smooth",
    metadata={"team" : "paris saint-germain"}
)

docs = [doc1, doc2, doc3]
```

```python

vector_store = Chroma (
    embedding_function= HuggingFaceEmbeddings(),
    persist_directory="./chroma_db",
    collection_name="football_players"
)
```

```python
vector_store.add_documents(docs)
```

```python
vector_store.get(include=["embeddings","metadatas", "documents"])
```

```python
vector_store.similarity_search(
    query = "who is best among all football players",
    k = 1 # how many simiar answers you want in output
)
```

```python
# meta-data filtering
vector_store.similarity_search_with_score(
    query = "",
    filter = {"team": "real madrid"}
)
```

```python
# update documents
updated_doc1 = Document(
    page_content="cristiano ronaldo is best football player in the world and he is the highest goal scorere in the history of football",
    metadata = {"team" : "al-nassar"}
)
vector_store.update_document(document_id = "e429b445-b8fb-4dfc-98be-edcdf40a4cc0",document = updated_doc1)
```
